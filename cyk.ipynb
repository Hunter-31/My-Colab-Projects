```python
import nltk
from nltk import CFG
from nltk.parse import RecursiveDescentParser, ShiftReduceParser, ChartParser

# -------------------------------
# 1) Define Grammar
# -------------------------------
grammar = CFG.fromstring("""
S -> NP VP
NP -> Det N | 'John' | 'Mary'
VP -> V NP
Det -> 'a' | 'the'
N -> 'dog' | 'cat'
V -> 'sees' | 'pets'
""")

sentence = "John sees a dog".split()

# -------------------------------
# 2) Top-Down Parsing
# -------------------------------
print("\nTop-Down Parsing:")
td_parser = RecursiveDescentParser(grammar)
for tree in td_parser.parse(sentence):
    print(tree)

# -------------------------------
# 3) Bottom-Up Parsing
# -------------------------------
print("\nBottom-Up Parsing:")
bu_parser = ShiftReduceParser(grammar)
for tree in bu_parser.parse(sentence):
    print(tree)

# -------------------------------
# 4) CYK Parsing (using ChartParser)
# -------------------------------
print("\nCYK Parsing (ChartParser):")
cyk_parser = ChartParser(grammar)
for tree in cyk_parser.parse(sentence):
    print(tree)
```
