```python
import re
from collections import defaultdict, Counter

# -------------------------------
# 1. Preprocess + Tokenize
# -------------------------------
def tokenize(text):
    # lowercase + keep words only + remove punctuations(.,!)
    tokens = re.findall(r'\b\w+\b', text.lower())
    return tokens

# -------------------------------
# 2. Build Unigram Model
# -------------------------------
def build_unigram(tokens):
    unigram_counts = Counter(tokens)
    total_tokens = len(tokens)
    vocab = set(tokens)
    V = len(vocab)
    return unigram_counts, total_tokens, V

def unigram_prob(word, unigram_counts, total_tokens, V):
    # Laplace smoothing
    return (unigram_counts[word] + 1) / (total_tokens + V)

# -------------------------------
# 3. Build Bigram Model
# -------------------------------
def build_bigram(tokens):
    bigram_counts = defaultdict(Counter)
    unigram_counts = Counter(tokens)

    for i in range(len(tokens) - 1):
        w1, w2 = tokens[i], tokens[i+1]
        bigram_counts[w1][w2] += 1

    vocab = set(tokens)
    V = len(vocab)

    return bigram_counts, unigram_counts, V

def bigram_prob(w1, w2, bigram_counts, unigram_counts, V):
    # Laplace smoothing
    return (bigram_counts[w1][w2] + 1) / (unigram_counts[w1] + V)

# -------------------------------
# 4. Example Text
# -------------------------------
text = "I love NLP and I love machine learning"

tokens = tokenize(text)
print("Tokens:", tokens)

# -------------------------------
# 5. Unigram Model
# -------------------------------
uni_counts, total, V = build_unigram(tokens)

print("\nUnigram Probabilities:")
for word in uni_counts:
    print(f"P({word}) = {unigram_prob(word, uni_counts, total, V):.4f}")

# -------------------------------
# 6. Bigram Model
# -------------------------------
bi_counts, uni_counts2, V2 = build_bigram(tokens)

print("\nBigram Probabilities:")
for w1 in bi_counts:
    for w2 in bi_counts[w1]:
        print(f"P({w2} | {w1}) = {bigram_prob(w1, w2, bi_counts, uni_counts2, V2):.4f}")
```
