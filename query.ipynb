```python
import re
import math
from collections import Counter, defaultdict

# -------------------------------
# 1. Sample Free Corpus (Gutenberg-style text)
# -------------------------------
documents = [
    "This is a sample document. It contains some common words and some rare words.",
    "This document is another example document. It also contains words.",
    "We are analyzing text data using TF and IDF. This is useful for search."
]

# -------------------------------
# 2. Stopwords (basic set)
# -------------------------------
stopwords = {
    'is','a','the','and','it','this','some','are','for','we','using','also'
}

# -------------------------------
# 3. Tokenization
# -------------------------------
def tokenize(text):
    return re.findall(r'\b\w+\b', text.lower())

# -------------------------------
# 4. Stopword Frequency Count
# -------------------------------
all_tokens = []
for doc in documents:
    all_tokens.extend(tokenize(doc))

stopword_freq = Counter([w for w in all_tokens if w in stopwords])

print("Stopword Frequencies:")
print(stopword_freq)

# Most frequent stopwords
print("\nMost Frequent Stopwords:")
print(stopword_freq.most_common(5))

# -------------------------------
# 5. TF Calculation
# -------------------------------
def compute_tf(doc_tokens):
    tf = Counter(doc_tokens)
    total = len(doc_tokens)
    return {word: count/total for word, count in tf.items()}

# -------------------------------
# 6. IDF Calculation
# -------------------------------
def compute_idf(docs):
    N = len(docs)
    idf = {}
    all_words = set(word for doc in docs for word in doc)

    for word in all_words:
        df = sum(1 for doc in docs if word in doc)
        idf[word] = math.log(N / df)

    return idf

# Prepare tokenized docs
tokenized_docs = [tokenize(doc) for doc in documents]

# Compute TF for each document
tf_list = [compute_tf(doc) for doc in tokenized_docs]

# Compute IDF
idf = compute_idf(tokenized_docs)

# -------------------------------
# 7. Query Analysis
# -------------------------------
query = "sample document words"
query_tokens = tokenize(query)

print("\nQuery Analysis (TF-IDF):")

for word in query_tokens:
    tf = tf_list[0].get(word, 0)   # using first document for example
    idf_val = idf.get(word, 0)
    tfidf = tf * idf_val
    print(f"{word}: TF={tf:.4f}, IDF={idf_val:.4f}, TF-IDF={tfidf:.4f}")
```
