```python
import math
from collections import Counter

docs = ["this is a sample document",
        "this document is another example"]

# Tokenize
docs = [d.lower().split() for d in docs]

# Vocabulary
vocab = set(w for doc in docs for w in doc)

# IDF
idf = {}
N = len(docs)
for w in vocab:
    df = sum(1 for doc in docs if w in doc)
    idf[w] = math.log(N / df)

# TF-IDF
def tfidf(doc):
    tf = Counter(doc)
    return [ (tf[w]/len(doc)) * idf[w] for w in vocab ]

v1 = tfidf(docs[0])
v2 = tfidf(docs[1])

# Cosine Similarity
dot = sum(a*b for a,b in zip(v1,v2))
mag1 = math.sqrt(sum(a*a for a in v1))
mag2 = math.sqrt(sum(b*b for b in v2))

print("Cosine Similarity:", dot/(mag1*mag2))
```
