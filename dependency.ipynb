```python
# Transition-Based Dependency Parser (Improved)

def parse(sentence):
    stack = []
    buffer = sentence.copy()
    arcs = []

    verbs = {"sees", "likes"}
    determiners = {"the", "a"}

    while buffer:
        # SHIFT first
        stack.append(buffer.pop(0))

        # Apply rules when possible
        while len(stack) >= 2:
            s1 = stack[-2]
            s2 = stack[-1]

            # Rule 1: determiner depends on noun
            if s1 in determiners:
                arcs.append((s2, s1))   # noun → det
                stack.pop(-2)

            # Rule 2: subject depends on verb
            elif s2 in verbs:
                arcs.append((s2, s1))   # verb → subject
                stack.pop(-2)

            # Rule 3: object depends on verb
            elif s1 in verbs and s2 not in determiners:
              arcs.append((s1, s2))   # verb → object
              stack.pop()

            else:
                break

    return arcs


# Test
sentence = ["the", "dog", "sees", "a", "cat"]

arcs = parse(sentence)

print("Sentence:", sentence)
print("\nCorrect Dependencies (head → dependent):")

for h, d in arcs:
    print(f"{h} → {d}")
```
