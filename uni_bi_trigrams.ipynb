```python
import re
from collections import Counter

# -------------------------------
# 1. Tokenization (sentence-wise)
# -------------------------------
def tokenize_sentences(text):
    sentences = re.split(r'[.!?]+', text)
    return [re.findall(r'\b\w+\b', s.lower()) for s in sentences if s.strip()]

# -------------------------------
# 2. Generate N-grams
# -------------------------------
def generate_ngrams(tokens, n):
    return list(zip(*[tokens[i:] for i in range(n)]))

# -------------------------------
# 3. Sample Corpus
# -------------------------------
corpus = """
This is my cat.
This is my dog.
My cat is cute.
"""

sentences = tokenize_sentences(corpus)

# Add sentence boundaries
all_tokens = []
for sent in sentences:
    tokens = ['<s>'] + sent + ['</s>']
    all_tokens.extend(tokens)

# -------------------------------
# 4. N-grams
# -------------------------------
unigrams = generate_ngrams(all_tokens, 1)
bigrams = generate_ngrams(all_tokens, 2)
trigrams = generate_ngrams(all_tokens, 3)

print("Unigrams:", unigrams)
print("\nBigrams:", bigrams)
print("\nTrigrams:", trigrams)

# -------------------------------
# 5. Counts
# -------------------------------
uni_counts = Counter(all_tokens)
bi_counts = Counter(bigrams)

V = len(set(all_tokens))  # vocabulary size

# -------------------------------
# 6. Sentence Probability
# -------------------------------
sentence = "This is my cat"
sent_tokens = ['<s>'] + re.findall(r'\b\w+\b', sentence.lower()) + ['</s>']

prob = 1

for i in range(len(sent_tokens) - 1):
    w1 = sent_tokens[i]
    w2 = sent_tokens[i+1]

    # Laplace smoothing
    numerator = bi_counts[(w1, w2)] + 1
    denominator = uni_counts[w1] + V

    prob *= numerator / denominator

print("\nSentence:", sentence)
print("Probability (Bigram with Laplace smoothing):", prob)
```
